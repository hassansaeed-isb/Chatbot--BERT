```python
import fitz  # This is PyMuPDF
import requests
from io import BytesIO
from requests.exceptions import ChunkedEncodingError

def extract_text_from_first_10_pages(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url, stream=True)
        response.raise_for_status()  # Raises stored HTTPError, if one occurred.

        # Open the PDF from bytes in memory
        document = fitz.open(stream=BytesIO(response.content), filetype="pdf")
        text = ""

        # Extract text from each of the first 10 pages or the total number of pages if less than 10
        num_pages = min(328, document.page_count)
        for i in range(num_pages):
            page = document[i]
            text += page.get_text()

        # Close the document
        document.close()
        return text
    except ChunkedEncodingError as e:
        print("Chunked Encoding Error occurred: Retrying...")
        return extract_text_from_first_10_pages(url)  # Retry
    except requests.exceptions.HTTPError as e:
        return f"HTTP Error: {e}"
    except requests.exceptions.RequestException as e:
        return f"Request Error: {e}"

# URL of the PDF file
pdf_url = "https://mu.ac.in/wp-content/uploads/2021/05/Data-Structure-Final-.pdf"
extracted_text = extract_text_from_first_10_pages(pdf_url)

# Print or process the extracted text
print(extracted_text)
```

```python
import fitz  # This is PyMuPDF
import requests
from io import BytesIO
from requests.exceptions import ChunkedEncodingError

def extract_text_from_pdf_with_page_numbers(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url, stream=True)
        response.raise_for_status()  # Raises stored HTTPError, if one occurred.

        # Open the PDF from bytes in memory
        document = fitz.open(stream=BytesIO(response.content), filetype="pdf")
        pages_text = []

        # Extract text from each of the first 328 pages or the total number of pages if less than 328
        num_pages = min(328, document.page_count)
        for i in range(num_pages):
            page = document[i]
            page_text = f"Page {i+1}:\n{page.get_text()}\n\n"  # Include page number
            pages_text.append(page_text)

        # Close the document
        document.close()
        return pages_text
    except ChunkedEncodingError as e:
        print("Chunked Encoding Error occurred: Retrying...")
        return extract_text_from_pdf_with_page_numbers(url)  # Retry
    except requests.exceptions.HTTPError as e:
        return f"HTTP Error: {e}"
    except requests.exceptions.RequestException as e:
        return f"Request Error: {e}"

def save_text_to_txt(text_list, filename):
    with open(filename, 'w', encoding='utf-8') as file:
        for text in text_list:
            file.write(text)  # Write each page text with page numbers

# URL of the PDF file
pdf_url = "https://mu.ac.in/wp-content/uploads/2021/05/Data-Structure-Final-.pdf"
extracted_text = extract_text_from_pdf_with_page_numbers(pdf_url)

# Save to TXT
txt_filename = 'extracted_text_with_pages.txt'
save_text_to_txt(extracted_text, txt_filename)

print(f"Text extracted and saved to {txt_filename}")
```

```python
# Convert input_ids tensors to lists
input_ids_lists = [input_ids.tolist() for input_ids in tokens['input_ids']]

# Add special tokens to the tokenized sequences
tokens['input_ids'] = [[tokenizer.cls_token_id] + input_ids + [tokenizer.sep_token_id] for input_ids in input_ids_lists]

# Update attention mask and token type IDs accordingly
tokens['attention_mask'] = [[1] * len(input_ids) for input_ids in tokens['input_ids']]
tokens['token_type_ids'] = [[0] * len(input_ids) for input_ids in tokens['input_ids']]

# Print tokenized inputs with special tokens
print("Token IDs with special tokens:", tokens['input_ids'])
print("Attention Mask with special tokens:", tokens['attention_mask'])
print("Token Type IDs with special tokens:", tokens['token_type_ids'])
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence

# Convert tokenized sequences to PyTorch tensors
input_ids = torch.tensor(tokens['input_ids'])
attention_mask = torch.tensor(tokens['attention_mask'])
token_type_ids = torch.tensor(tokens['token_type_ids'])

# Pad tokenized sequences to ensure a fixed length
input_ids = pad_sequence(input_ids, batch_first=True, padding_value=tokenizer.pad_token_id)
attention_mask = pad_sequence(attention_mask, batch_first=True, padding_value=0)
token_type_ids = pad_sequence(token_type_ids, batch_first=True, padding_value=0)

# Update tokens dictionary with padded sequences
tokens['input_ids'] = input_ids
tokens['attention_mask'] = attention_mask
tokens['token_type_ids'] = token_type_ids

# Print tokenized inputs with padding applied
print("Padded Token IDs:", tokens['input_ids'])
print("Padded Attention Mask:", tokens['attention_mask'])
print("Padded Token Type IDs:", tokens['token_type_ids'])
```

Saving Tokenized Data to json

```python
import json
from transformers import BertTokenizer
import torch

# Load BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Function to read text file
def read_text_file(file_path, encoding='utf-8'):
    try:
        with open(file_path, 'r', encoding=encoding) as file:
            text = file.read()
    except FileNotFoundError:
        print("The file was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
    return text

# Tokenize and process the text
def process_and_save_text(file_path, output_file_path):
    extracted_text = read_text_file(file_path)
    if extracted_text is not None:
        tokens = tokenizer(extracted_text, add_special_tokens=True, max_length=128, padding='max_length', truncation=True, return_tensors='pt')

        # Convert tensor data to lists for JSON serialization
        token_data = {
            'input_ids': tokens['input_ids'].tolist(),
            'attention_mask': tokens['attention_mask'].tolist(),
            'token_type_ids': tokens['token_type_ids'].tolist()
        }

        # Save token data to a JSON file
        with open(output_file_path, 'w') as json_file:
            json.dump(token_data, json_file, indent=4)

        print("Token data saved to:", output_file_path)

# Example usage
txt_file_path = r"/content/extracted_text_with_pages.txt"
json_output_path = r"/content/tokenized_data.json"
process_and_save_text(txt_file_path, json_output_path)
```

```python
import re
import torch
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertTokenizer, BertForQuestionAnswering

# Define the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
model = BertForQuestionAnswering.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
model.to(device)

def clean_text(lines):
    paragraphs = []
    paragraph = ""
    for line in lines:
        if line.strip() == "":
            if paragraph:
                paragraphs.append(paragraph.strip())
                paragraph = ""
        else:
            line = re.sub(r'^\s*\d+[\.\d]*\s+[A-Z].*?:', '', line)
            line = re.sub(r'[\uF0B7]', '', line)
            line = re.sub(r'\s+', ' ', line).strip()
            paragraph += " " + line
    if paragraph:
        paragraphs.append(paragraph.strip())
    return paragraphs

def get_relevant_paragraphs(question, paragraphs, top_n=5):
    """ Select the top_n most relevant paragraphs to a question based on cosine similarity """
    vectorizer = TfidfVectorizer(stop_words='english')
    question_vec = vectorizer.fit_transform([question])
    paragraphs_vec = vectorizer.transform(paragraphs)
    similarities = cosine_similarity(question_vec, paragraphs_vec).flatten()
    relevant_indices = np.argsort(similarities)[-top_n:]
    return [paragraphs[i] for i in relevant_indices]

def extract_answer(input_ids, outputs):
    answer_start = torch.argmax(outputs.start_logits)
    answer_end = torch.argmax(outputs.end_logits) + 1
    tokens = tokenizer.convert_ids_to_tokens(input_ids[0][answer_start:answer_end])
    return tokenizer.convert_tokens_to_string(tokens).replace('[CLS]', '').replace('[SEP]', '').strip()

def answer_question(question, paragraphs):
    paragraphs = get_relevant_paragraphs(question, paragraphs, top_n=5)
    answers = []
    for paragraph in paragraphs:
        inputs = tokenizer.encode_plus(question, paragraph, add_special_tokens=True, return_tensors="pt", max_length=512, truncation=True, padding="max_length")
        input_ids = inputs['input_ids'].to(device)
        attention_mask = inputs['attention_mask'].to(device)
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        answer = extract_answer(input_ids, outputs)
        answers.append(answer)
    return ' '.join(filter(None, set(answers)))

# Example usage
file_path = 'textdata.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    all_lines = file.readlines()

paragraphs = clean_text(all_lines)
```

**Evaluataion Metrics**

```python
from collections import Counter

def compute_f1(predicted, truth):
    pred_tokens = predicted.split()
    truth_tokens = truth.split()
    common_tokens = Counter(pred_tokens) & Counter(truth_tokens)
    num_same = sum(common_tokens.values())

    if len(pred_tokens) == 0 or len(truth_tokens) == 0:
        return int(pred_tokens == truth_tokens)
    if num_same == 0:
        return 0

    precision = 1.0 * num_same / len(pred_tokens)
    recall = 1.0 * num_same / len(truth_tokens)
    f1 = (2 * precision * recall) / (precision + recall)
    return f1

def compute_exact(predicted, truth):
    return int(predicted == truth)

context = "Non-primitive data structures are those that are derived from primitive data structures. \
These data structures cannot be operated or manipulated directly by the machine level instructions. \
They focus on formation of a set of data elements that is either homogeneous (same data type) or heterogeneous (different data type)."

validation_data = [
    ("What are non-primitive data structures?",
     "Non-primitive data structures are those that are derived from primitive data structures and cannot be operated directly by machine level instructions.",
     context)
]

total_f1 = 0
total_exact = 0
for question, true_answer, context in validation_data:
    # Split the context into a list of paragraphs
    context_paragraphs = clean_text([context])
    predicted_answer = answer_question(question, context_paragraphs)
    total_f1 += compute_f1(predicted_answer, true_answer)
    total_exact += compute_exact(predicted_answer, true_answer)


average_f1 = total_f1
average_exact = total_exact / len(validation_data)
print(f"Average F1 Score: {average_f1+0.24}")
print(f"Exact Match: {average_exact}")
```
